## Datasets 

### The 3 data sets are as follows. They contain the prices and volumes of imported crude oil from 1973 to 2025 and are split into the persian gulf, OPEC countries and non opec countries}

```python
import pandas as pd 
import re
gulf=pd.read_csv('Perisan Gulf Crude import volume .csv')
opec=pd.read_csv('Opec COuntries crude import volumer.csv')
nonopec=pd.read_csv('non opecs.csv')
```

### Cleaning the OPEC dataframe


```python
x=opec.columns
```

```python
x=list(x)
x
```

```python
titles =x
country_names = []

for title in titles:
    match = re.search(r'from (.+?) of Crude Oil', title)
    if match:
        country_names.append(match.group(1))

print(country_names)
```

```python
country_names.insert(0,'Date')
```

```python
country_names
```

```python
opec.columns=country_names
```

```python
opec
```
